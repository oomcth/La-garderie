```python
audio_file_path = 'azisu.wav'
wav_fpath = Path(audio_file_path)
file_duration = 194

wav = preprocess_wav(wav_fpath)

encoder = VoiceEncoder("cpu")
_, cont_embeds, wav_splits = encoder.embed_utterance(wav, return_partials=True, rate=16)
print(cont_embeds.shape)
```

```python
time_to_index = len(cont_embeds)/file_duration
beginning = 30
end = 60
```

```python
ms = cluster.MeanShift(cluster_all = False)
ms.fit(cont_embeds[int(beginning*time_to_index):int(end*time_to_index)])
#print(ms.cluster_centers_)
print(ms.cluster_centers_.shape)
```

```python
from sklearn.cluster import SpectralClustering
sc = SpectralClustering()
sc.fit(cont_embeds)
```

```python
T = np.linspace(0, file_duration, len(cont_embeds))
#np.arange(len(cont_embeds))*#int(beginning*time_to_index),int(end*time_to_index))/time_to_index
plt.plot(T, sc.labels_)
plt.show()
```
