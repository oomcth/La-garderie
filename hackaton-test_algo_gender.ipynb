```python
#on va chercher les audios
from tqdm import tqdm
liste_chemins=list(data["chemin"])
wavs = []
for wav_fpath in tqdm(liste_chemins):
    try:
       r = preprocess_wav("/content/" + wav_fpath)
    except Exception as e:
        print(wav_fpath, e)
        continue
    wavs.append(r)
```

```python
encoder = VoiceEncoder()
utterance_embeds = np.array(list(map(encoder.embed_utterance, wavs)))
```

```python
len(utterance_embeds)
```

```python
data["utterance_embeds"]=list(utterance_embeds)
data
```

```python
#on transformance l'utterance embeds en features
x=[0 for element in range(256)]
for i in range(256):
  x[i]=[]

for element in data["utterance_embeds"]:
  for i in range(256):
    x[i].append(element[i])
```

```python
columns_names=[f"x{i}" for i in range(256)]
df_features = pd.DataFrame(np.array(x).T, columns=columns_names)
df_features
```

```python
df_features["sound_clip_id"]=data["sound_clip_id"]
```

```python
df_with_features = pd.merge(data, df_features, on="sound_clip_id")
df_with_features.head()
```

```python
import sklearn.model_selection
from sklearn.linear_model import LogisticRegression

from sklearn import svm
from sklearn.model_selection import cross_val_score
```

```python
# regression logistique

X=df_with_features.iloc[:,17:]
y=df_with_features["nb_gender_M"]

modele_logit = LogisticRegression(penalty='none',solver='newton-cg')
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2)
```

```python
def erreur(results, donnees) : 
  nbre_erreurs=0
  for i in range(len(results)):
    if results[i]!=donnees[i]:
      nbre_erreurs+=1
  #print(nbre_erreurs)
  train_error=nbre_erreurs/len(results)
  return train_error
```

```python
clf = svm.SVC(kernel='linear', random_state=42)
clf.fit(X_train,y_train)
```

```python
results_train=clf.predict(X_train)
results_test=clf.predict(X_test)

print("Erreur d'entrainement: ",100*erreur(results_train, np.array(y_train)),"%")
print("Erreur de test: ",100*erreur(results_test, np.array(y_test)),"%")
```

```python
def predict(path):
  wav = preprocess_wav(path)
  encoder = VoiceEncoder()
  embed = encoder.embed_utterance(wav)
  feature=embed=embed.reshape(-1,1).T
  prediction=clf.predict(feature)
  return prediction[0]
```
